What Kind of Data to Use?
Perfect for:
1. Two continuous numerical variables
2. Finding correlations
    sns.jointplot(data=df, x='study_hours', y='exam_score', kind='reg')
#Shows if more study = higher score
3. Detecting outliers
    sns.jointplot(data=df, x='age', y='income', kind='scatter')
#Unusual points are easy to spot
4. Understanding distributions together
    sns.jointplot(data=df, x='temperature', y='sales', kind='kde')
#See how both variables distribute and relate
NOT suitable for:
❌ Categorical variables (use catplot or countplot instead)
❌ More than 2 variables (use pairplot for multiple variables)
❌ Time series (use lineplot instead)
❌ Single variable (use histplot or kdeplot)

Tips and Best Practices

Use kind='hex' or kind='kde' for large datasets (>1000 points)
Use kind='reg' to see trend lines and correlations
Use hue to compare groups (requires seaborn >= 0.11)
Adjust ratio parameter to make marginal plots larger/smaller
Add transparency with alpha for overlapping points
